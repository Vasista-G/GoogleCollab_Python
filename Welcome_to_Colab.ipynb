```
def show_balance():
  print(f"Your balance is {balance}")

def withdraw():
  global balance
  amount = int(input("Enter amout to withdraw :"))

  if amount > 0:
    print("Insufficient funds, blance remaining is ", balance)
  else:
    balance = balance - amount
    print("Amount withdraw is successfull.")
    print("Remaining Balance is ", balance)

def deposit():
  global balance
  amount = int(input("Enter amount to deposit :"))

  if amount <= 0:
    print("Invalid amount")
  else:
    balance = balance + amount
    print("Amount deposit is successfull.")
    print("Remaining Balance is ", balance)
```

Python magic methods

```
class Book:

  def __init__(self,name, author,pages):
    self.name = name
    self.author = author
    self.pages = pages

  def __str__(self) -> str:
    return f"Book name {self.name}, Pages {self.pages}"

  def __add__(self, other):
    return Book(self.name, self.author, self.pages + other.pages)

book1 = Book("Python", "Kaisen",50)
book2 = Book("Java", "Lolly",30)

print(book1)
print(book2)

print(book1 + book2)
```

@property is a built-in Python decorator that lets you define methods that can be accessed like attributes. It is commonly used to implement encapsulation, validation, computed values, and read-only attributes while keeping a clean interface.  
A property can have three methods:  
1. getter  
2. setter  
3. deleter  


```
class Student:

    def __init__(self, name, age):
        self._name = name
        self._age = age


    #Getter name
    @property
    def name(self):
        print("inside name Getter")
        return self._name

    #Getter age
    @property
    def age(self):
        print("Inside age Getter")
        return self._age

    #Setter age
    @age.setter
    def age(self, value):
        print("inside age setter")

        if value < 0:
            raise ValueError("Age cannot be negative")

        self._age = value

s = Student("Rahul", 30)

print(s.name + '\n')
# print(s.age)

s.age = 20

print(s.age)
```
